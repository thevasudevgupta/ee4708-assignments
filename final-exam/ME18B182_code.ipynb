### 2. Sector-wise Stock analysis

#### 2.1 ICICI, HDFC and SBI (Banking sector)

```python
ICICI['Open'].plot(label="ICICI Open", figsize=(15,7))
HDFC['Open'].plot(label="HDFC Open")
SBI['Open'].plot(label="SBI Open")
plt.ylabel('Stock Price in INR')
plt.legend()
```

```python
ICICI['Volume'].plot(label="ICICI Volume", figsize=(15,7))
HDFC['Volume'].plot(label="HDFC Volume")
SBI['Volume'].plot(label="SBI Volume")
plt.ylabel('Volume traded')
plt.legend()
```

We can try to get an idea about the total money involved in trading by calculating ``Total Traded`` (an indicator; not true value) = ``Open`` * ``Volume``.

```python
ICICI['Total traded'] = ICICI['Open']*ICICI['Volume']
HDFC['Total traded'] = HDFC['Open']*HDFC['Volume']
SBI['Total traded'] = SBI['Open']*SBI['Volume']
```

```python
ICICI['Total traded'].plot(label="ICICI total traded", figsize=(15,7))
HDFC['Total traded'].plot(label="HDFC total traded")
SBI['Total traded'].plot(label="SBI total traded")
plt.ylabel('Total traded')
plt.legend()
```

### Daily %tage change
$r_t=\frac{p_t}{p_{t-1}}-1$

```python
ICICI['Returns'] = ICICI['Close']/ICICI['Close'].shift(1) - 1
HDFC['Returns'] = HDFC['Close']/HDFC['Close'].shift(1) - 1
SBI['Returns'] = SBI['Close']/SBI['Close'].shift(1) - 1
```

```python
ICICI['Returns'].plot(kind='kde', label='ICICI', figsize=(10,8))
HDFC['Returns'].plot(kind='kde', label='HDFC')
SBI['Returns'].plot(kind='kde', label='SBI')
plt.legend()
```
